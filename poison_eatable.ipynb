## Packages

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Load data

```python
data = pd.read_csv("/home/rohullah/Desktop/repo/data/mushroom/mushrooms.csv")
```

```python
data.head(10)
```

## Distributions

```python
plt.figure(figsize=(8,6))
sns.countplot(data=data, x='class', hue='class')
plt.legend(['P = Poisonous', 'E = Edible'])
plt.title("Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=data, x='cap-shape', hue='class')
plt.legend(['Bell = b','Conical = c','Convex=x','Flat=f', 'Knobbed=k,Sunken=s'])
plt.title("Shape Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=data, x='cap-surface', hue='class')
plt.legend(['Fibrous = f','gGrooves = g','Scaly=y','Smooth=s'])
plt.title("Cap Surface")
plt.show()
```
